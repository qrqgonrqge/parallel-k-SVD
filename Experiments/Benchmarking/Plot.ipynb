```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("omp_benchmark_all.csv")
functions = df["function"].unique()
```

```python
functions
```

```python
plt.figure()
for fn in functions:
    subset_fn = df[df["function"] == fn]
    threads_list = sorted(subset_fn["threads"].unique())
    # sub = subset_fn[subset_fn["threads"] == t]
    sub = subset_fn
    plt.plot(
        sub["batch_size"],
        sub["time_sec"],
        marker='o',
        label=f"{fn}"
    )
    plt.xscale("log", base=2)
    plt.yscale("log")
    plt.xlabel("Batch Size")
    plt.ylabel("Time (s)")
    plt.title('OMP Showdown')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
plt.show()
```
